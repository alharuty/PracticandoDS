```
# Load data
import pandas as pd

dataset = pd.read_csv('adult.csv')
```

```
# Your answers here!

import pandas as pd
import matplotlib.pyplot as plt

# Crear una nueva columna para agrupar por rangos
bins = list(range(0, 101, 10))  # 0-9, 10-19, ..., 90-99
labels = [f'{i}-{i+9}' for i in bins[:-1]]
dataset['age_group'] = pd.cut(dataset['age'], bins=bins, labels=labels, right=False)

# Contar cuántos hay en cada grupo
age_counts = dataset['age_group'].value_counts().sort_index()

# Graficar
plt.figure(figsize=(12, 5))
bars = plt.bar(age_counts.index.astype(str), age_counts.values, color='#4C72B0', edgecolor='black')

# Títulos y etiquetas
plt.title("Distribución de edades por rangos de 10 años", fontsize=16)
plt.xlabel("Rango de edad", fontsize=12)
plt.ylabel("Cantidad de personas", fontsize=12)

# Mostrar los valores arriba de las barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, str(int(height)), ha='center', va='bottom', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
